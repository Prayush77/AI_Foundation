```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#Line plot
x = np.linspace(0,10,100)
y = np.tan(x)
```

```python
# print(x)
# print('__________')
# print(y)
```

```python
plt.plot(x,y,label ="Sinusoidal wave")
plt.title("Line plot example graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
#Barchart
categories = ["A","B","C","D"]
values = [10,20,15,25]
plt.bar("A",10,color="pink")
plt.bar("C",20,color="red")
plt.bar("B",15,color="pink")
plt.bar("D",25,color="blue")

plt.title("Bar chart example graph")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#"Histogram" and KDE plot
data = np.random.randn(1000 )
sns.histplot(data,kde=True,color="Green")
plt.title("Histogram and KDE plot example graph")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Heat map # used in correlation checking accuracy in model

matrix = np.random.rand(5,5)
sns.heatmap(matrix,annot=True,cmap="viridis",fmt=".2f")
plt.title("Heat map example graph")
plt.show()
```
